```python
import pandas as pd
from nltk.stem import PorterStemmer as st
from textblob import TextBlob
from textblob import Word
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('vader_lexicon')
```

```python
def data_pre (df):
    # Adds a column with the word count
    df['WORD_COUNT'] = df['AD_TEXT'].apply(lambda x: len(str(x).split(" ")))
    
    # Removes all entries, where the ad text is less than 5 words
    df = df[df['WORD_COUNT'] > 4]
    
    # removes hyperlinks
    df['AD_TEXT']=df['AD_TEXT'].str.replace(r'https?:\/\/.*[\r\n]*', '')
    
    # removes stopwords from ad text, after cloning the ad text column
    stop = stopwords.words('english')
    df['AD_CLEANED_TEXT']=df['AD_TEXT']
    df['AD_CLEANED_TEXT'].apply(lambda x: " ".join(x for x in str(x).split() if x not in stop))
    
    # Sets the words to lowercase
    df['AD_CLEANED_TEXT'] = df['AD_CLEANED_TEXT'].apply(lambda x: " ".join(x.lower() for x in str(x).split()))
    
    # removes punctuation
    df['AD_CLEANED_TEXT']=df['AD_CLEANED_TEXT'].str.replace('[^\w\s]','')
    df['AD_CLEANED_TEXT']=df['AD_CLEANED_TEXT'].str.replace("[^a-zA-Z#]", " ")
    
    # Stemming of the ad text
    st = PorterStemmer()
    df['AD_CLEANED_TEXT'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
    
    # Lemmatization - converts the word into its root word
    df['AD_CLEANED_TEXT'].apply(lambda x: " ".join([Word(word).lemmatize() for word in str(x).split()]))
    
    return df
```
